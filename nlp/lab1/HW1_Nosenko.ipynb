```python
text = " ".join(text5)
text_corrected = " ".join(word for word in text7 if word.isalpha())
word_list = nltk.word_tokenize(text_corrected)
```

```python
freq_dist = FreqDist(word_list)
hapaxes = freq_dist.hapaxes()
low_freq_words = [word for word, freq in freq_dist.items() if freq < 5]
print("Слова-унікаліними (hapaxes):", hapaxes)
print("Слова, які зустрічаються менше 5 разів:", low_freq_words)
```

```python
word_lengths = [len(word) for word in word_list]

length_freq_dist = FreqDist(word_lengths)

sorted_keys = sorted(length_freq_dist.keys())
sorted_values = [length_freq_dist[key] for key in sorted_keys]

plt.plot(sorted_keys, sorted_values)
plt.xticks(range(min(sorted_keys), max(sorted_keys)+1))
plt.xlabel('Довжина слова')
plt.ylabel('Частота')
plt.title('Графік довжини слів до їх частоти (відсортований)')

plt.show()
```

```python
import re

def sum_monetary_amounts(tokens):
    # Регулярний вираз для виявлення грошових сум
    pattern = r'\$\d+\.\d{0,1}'

    # Знаходимо всі входження грошових сум у тексті
    matches = re.findall(pattern, ' '.join(tokens))

    # Конвертуємо входження у числа з плаваючою точкою
    amounts = [float(match[1:]) for match in matches]

    total = sum(amounts)
    return total


with open('It.txt', 'r') as file:
    text = file.read()
    tokens = nltk.sent_tokenize(text)

total_amount = sum_monetary_amounts(tokens)

print("Загальна сума грошових виплат:", total_amount)
```

```python
import json
import re

initial_data = {
    "HTTPHost": "localhost",
    "HTTPPort": 8545,
    "Config": {
        "Enabled": True,
        "Server": "wakuv2.prod",
        "Nodes": [
            "8.210.222.231",
            "168.166.125.145",
            "192.168.0.1",
            "8.210.222.234",
            "8.210.222.588"
        ]
    }
}


with open('initial_data.json', 'w') as file:
    json.dump(initial_data, file, indent=4)

with open('initial_data.json', 'r') as file:
    data = json.load(file)

data['Config']['Nodes'] = [node for node in data['Config']['Nodes'] if not re.match(r'^8\.', node)]

server_name, server_type = re.match(r'(\w+)\.(\w+)', data['Config']['Server']).groups()
data['Config']['Server'] = {"name": server_name, "type": server_type}

with open('edited_data.json', 'w') as file:
    json.dump(data, file, indent=4)
```
